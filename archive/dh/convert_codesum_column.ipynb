```python
def fill_value_codesum(codesum):
    if codesum == " ":
        return "MO"
    else:
        return codesum
```

```python
def ch_codesum(codesum):
    result = ""
    
    codesum = codesum.replace(" ","")
    codesum = codesum.replace("+","")
    
    for s in range(0,len(codesum)):
        if s % 2 == 0 and s != 0:
            result += ',' + codesum[s]
        else:
            result += codesum[s]
    return result
```

```python
data['codesum'] = data['codesum'].apply(ch_codesum)
```
