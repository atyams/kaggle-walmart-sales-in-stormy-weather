# ols

```python
df_test.columns
```

```python
features = ['date', 'store_nbr', 'item_nbr', 'units', 'station_nbr', 'tmax', 'tmin',
       'tavg', 'depart', 'dewpoint', 'wetbulb', 'heat', 'cool', 'sunrise',
       'sunset', 'codesum', 'codesum_YN', 'snowfall', 'preciptotal',
       'stnpressure', 'sealevel', 'resultspeed', 'resultdir', 'avgspeed',
       'dayofweek', 'is_holiday']
features = " ".join(features)
features = features.replace(" ", " + ")
features
```

```python
model_1 = sm.OLS.from_formula('units ~ C(item_nbr) + tmax + tmin + tavg + depart + dewpoint + wetbulb + heat + cool + sunrise + sunset + C(codesum_YN) + snowfall + preciptotal + stnpressure + sealevel + resultspeed + resultdir + avgspeed + C(is_holiday) + 0', data=df_test)
result_1 = model_1.fit()
print(result_1.summary())
```

# `Rsquared, Cond. No. , Durbin-Watson`

# submission preparation

```python
test = pd.read_csv('../data/basic/test.csv')
print(test.shape)
test.head()
```

```python
test = pd.merge(test, key)
```

```python
weather_basic = pd.read_csv('../data/basic/weather.csv')
```

```python
test = pd.merge(test, weather_basic)
```

```python
test = pd.DataFrame(test, columns=['date', 'tmax', 'tmin', 'tavg',
       'depart', 'dewpoint', 'wetbulb', 'heat', 'cool', 'sunrise', 'sunset',
       'codesum', 'snowfall', 'preciptotal', 'stnpressure', 'sealevel',
       'resultspeed', 'resultdir', 'avgspeed', 'station_nbr', 'store_nbr', 'item_nbr'])
```

```python
test.head()
```

```python
for i in test.columns:
    test[i] = test[i].replace('-', np.nan)
    test[i] = test[i].replace(' ', np.nan)
    test[i] = test[i].replace('M', np.nan)
    test[i] = test[i].replace('T', np.nan)
    test[i] = test[i].replace('  T', np.nan)
```

```python
for col in test[['tmax', 'tmin', 'tavg', 'depart', 'dewpoint', 'wetbulb', 'heat',
       'cool', 'sunrise', 'sunset', 'snowfall', 'preciptotal',
       'stnpressure', 'sealevel', 'resultspeed', 'resultdir', 'avgspeed',
       'station_nbr', 'store_nbr', 'item_nbr']]:
    test[col] = test[col].astype(np.float64)
```

```python
test.columns
```

```python
test_sample = test.drop(columns=['date', 'tmax', 'depart', 'cool','preciptotal', 'resultspeed', 'resultdir', 'codesum', 'station_nbr', 'store_nbr'])
```

```python
test_sample['item_nbr'] = test_sample['item_nbr'].astype(int)
```

```python
test_sample
```

```python
use_item_nbr = df1['item_nbr'].unique()
use_item_nbr.sort()
use_item_nbr
```

```python
test_sample_1 = test_sample[test_sample['item_nbr'].isin(use_item_nbr)]
test_sample_2 = test_sample[test_sample['item_nbr'].isin(use_item_nbr) == False]
```

# Predict

codesum 컬럼에서 에러가 떴습니다.. 이 부분은 따로 제출을 위한 encoding이 필요해보입니다. 여기서는 그냥 뺐습니다.

```python
test_result = result_1.predict(test_sample_1)
```

```python
test_result.describe()
```

# preprocess for submission

```python
result = pd.concat([test, test_result], axis=1)
```

```python
print(result.shape)
result.head()
```

```python
result['units'] = result[0]
```

```python
result = pd.DataFrame(data=result, columns=['store_nbr', 'item_nbr', 'date', 'units'])
```

```python
result.head()
```

```python
for col in result[['store_nbr', 'item_nbr']]:
    result[col] = result[col].astype(int)
```

```python
result['id'] = result['store_nbr'].astype(str)+'_'+result['item_nbr'].astype(str)+'_'+result['date'].astype(str)
```

```python
result.head()
```

```python
result = result.replace(np.nan, 0)
```

```python
result = pd.DataFrame(data=result, columns=['id','units'], index = None)
```

```python
result['units'] = result['units'].apply(np.ceil)
```

```python
print(result.shape)
result.head()
```

```python
result.to_csv('submission_sample.csv', index=False)
```
