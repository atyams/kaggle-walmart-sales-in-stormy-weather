### rename codesum 
- blank -> plain weather
- codesum -> abnormal weather (in `Cdsm_Ch` column)

```python
def ch_codesum(codesum):
    if codesum == ' ':
        return "Plain"
    else:
        return codesum
```

```python
data['codesum'] = data['codesum'].apply(ch_codesum)
```

```python
def compare_codesum(codesum):
    if codesum =="Plain":
        return codesum
    else:
        return "Abnormal"
```

```python
data["Cdsm_Ch"] = data['codesum'].apply(compare_codesum)
```

### Rearrange 'date'

```python
def convert_day(day):
    if day == 'Monday':
        return 'Mon'
    elif day == 'Tuesday':
        return 'Tue'
    elif day == 'Wednesday':
        return 'Wed'
    elif day == 'Thursday':
        return 'Thu'
    elif day == 'Friday':
        return 'Fri'
    elif day == 'Saturday':
        return 'Sat'
    elif day == 'Sunday':
        return 'Sat'
```

```python
holiday['dayofweek'] = holiday['day'].apply(convert_day)
```

```python
holiday = holiday.drop('day', axis=1)
```

```python
holiday.tail()
```

```python
def apply_day(date):
    return date[-2:]
```

```python
holiday['day'] = holiday['date'].apply(apply_day)
```

```python
holiday
```

```python
def apply_month(date):
    return date[:-2]
```

```python
holiday['month'] = holiday['date'].apply(apply_month)
```

```python
def convert_month(month):
    if month == 'January':
        return '01'
    elif month == 'February':
        return '02'
    elif month == 'March':
        return '03'
    elif month == 'April':
        return '04'
    elif month == 'May':
        return '05'
    elif month == 'June':
        return '06'
    elif month == 'July':
        return '07'
    elif month == 'August':
        return '08'
    elif month == 'September':
        return '09'
    elif month == 'October':
        return '10'
    elif month == 'November':
        return '11'
    elif month == 'December':
        return '12'
```

```python
holiday['month'] = holiday['month'].apply(convert_month)
```

```python
holiday = holiday.drop('date', axis = 1)
```

```python
holiday['date'] = holiday['year'].astype(str) + '-' + holiday['month'].astype(str) + '-' + holiday['day'].astype(str) 
```
