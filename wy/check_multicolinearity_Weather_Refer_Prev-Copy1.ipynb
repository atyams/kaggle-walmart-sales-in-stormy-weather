```python
model_weather = sm.OLS.from_formula("tmax ~ tmin + tavg + depart + dewpoint + wetbulb + heat + cool + sunrise + sunset + snowfall + preciptotal + stnpressure + sealevel + resultspeed + resultdir + avgspeed", data=weather)
result_weather = model_weather.fit()
sm.stats.anova_lm(result_weather, typ=2)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(weather2.values, i) for i in range(weather2.shape[1])]
vif["features"] = weather2.columns
```

```python
vif.sort_values(by=['VIF Factor'], ascending=False).reset_index(drop=True)
```
