# 1. load .csv file

```python
data1 = pd.read_csv('../data/weather_v1.csv', index_col = 0)
data2 = pd.read_csv('../data/weather_v2.csv', index_col = 0)
data3 = pd.read_csv('../data/weather_v3.csv', index_col = 0)
```

# 2. data컬럼의 년, 월, 일 분리하는 함수

```python
def make_year(date):
    return int(date.split('-')[0])

def make_month(date):
    return int(date.split('-')[1])

def make_day(date):
    return int(date.split('-')[2])

def make_monthday(date):
    return (date.split('-')[1:3])
```

# 3. 총 일수 365, 366 일 구하는 함수

```python
def is_leap_year(year):
    return ((year % 4 == 0 and year % 100 != 0) or year % 400 == 0)

def sum_of_days(year,monthday):
    
    month = int(monthday[0])
    day = int(monthday[1])
    
    common_year = [
        31,
        28 + + int(is_leap_year(year)),
        31,
        30,
        31,
        30,
        31,
        31,
        30,
        31,
        30,
        31
    ]
    tmp_days = 0
    for i in range(month - 1):
        tmp_days += common_year[i]
    tmp_days += day
    return tmp_days
```

# 4. 구한 요일 수를 통해 요일을 구할 수 있다.

```python
# 요일 구하는 함수
def check_week_of_day(year,days):
    
    if year == 2012:

        date_list = ['Sun', 'Mon', 'Tue', 'Wed', 'Thi', 'Fri', 'Sat']
        
    elif year == 2013:
        
        date_list = ['Tue', 'Wed', 'Thi', 'Fri', 'Sat', 'Sun', 'Mon']
    else:
        
        date_list = ['Wed', 'Thi', 'Fri', 'Sat', 'Sun', 'Mon','Tue']
    
    return date_list[days % 7 - 1]
```

# 5. 각 버전에 함수 적용!!

```python
data1['year'] = data1['date'].apply(make_year)
data1['monthday'] = data1['date'].apply(make_monthday)
data1['days'] = data1.apply(lambda row : sum_of_days(row['year'], row['monthday']), axis = 1)
data1['dayofweek'] = data1.apply(lambda row : check_week_of_day(row['year'], row['days']), axis = 1)
```

# 6. 주중인데 휴일인 경우를 고려하면

```python
holiday = pd.read_csv('../wy/holiday',index_col = 0)
```

```python
holiday
```

### data의 리스트만 뽑아서 리스트를 만들고

```python
holiday_list = list(holiday['date'])
```

```python
def is_holiday(date, dayofweek):
    if date in holiday_list:
        return 'holiday_work'
    elif dayofweek == "Sun" or dayofweek == "Sat":
        return 'holiday'
    else:
        return 'workday'
```

# 7. is_holiday 함수 적용 

```python
data1['is_holiday'] = data1.apply(lambda row : is_holiday(row['date'], row['dayofweek']), axis = 1)
data2['is_holiday'] = data2.apply(lambda row : is_holiday(row['date'], row['dayofweek']), axis = 1)
data3['is_holiday'] = data3.apply(lambda row : is_holiday(row['date'], row['dayofweek']), axis = 1)
```
