# sunrise, sunset conversion to numerical value

```python
def converge_sun(sun):
    if sun != '-':
        return float(sun[:2]) * 60 + float(sun[2:4])
```

```python
weather['sunrise'] = weather['sunrise'].apply(converge_sun)
weather['sunset'] = weather['sunset'].apply(converge_sun)
```

# replace missing value with np.nan

```python
for i in weather.columns:
    print(i, "\n", weather[i].unique())
```

```python
def codesum_character(codesum):
    if codesum == ' ':
        return "moderate"
    else:
        return codesum
```

```python
weather['codesum'] = weather['codesum'].apply(codesum_character)
```

```python
def compare_codesum(codesum):
    if codesum =="moderate":
        return codesum
    else:
        return "abnormal"
```

```python
weather['codsum_YN'] = weather['codesum'].apply(compare_codesum)
```

```python
for i in weather.columns:
    weather[i] = weather[i].replace('-', np.nan)
    weather[i] = weather[i].replace(' ', np.nan)
    weather[i] = weather[i].replace('M', np.nan)
    weather[i] = weather[i].replace('T', np.nan)
    weather[i] = weather[i].replace('  T', np.nan)
```

```python
for i in weather.columns:
    print(i, "\n", weather[i].unique())
```

## Missing Data를 채워넣는 방법 고민 (수정)



- `tmax` = 전 날 참조 
- `tmin` = 전 날 참조

- `depart` = 첫 row는 0 으로 채워넣은 후, 전 날 참조
        - 지난 30년의 평균값 - `tmax` 값이므로, '현재로서는 평균값과 차이가 없다' 라는 주장이 제일 오차를 줄일 수 있을 것 같다.

- `wetbulb 습도` = 전 날 참조
- `dewpoint 이슬점` = 전 날 참조


- `sunrise-sunset` = 전 날 참조

- `preciptotal` = 전 날을 참조

- `station pressure 기압` = 전 날 참조
- `sealevel 해면기압` = 전 날 참조 
- `resultdir` = 전 날 참조
- `avgspeed` = 전 날 참조



- `tavg` = (`tmax` + `tmin` / 2) 반올림
- `heat` = 65 - `tavg`   (난방 지수)
- `cool` = `tavg` + 65   (냉방 지수)
        - 기준 온도 화씨 65도 (18도) (참조 : https://www.wunderground.com/about/faq/degreedays.asp)
        
- `snowfall` = preciptotal의 값을 참조

### 전 날 참조 함수

```python
def fill_prev():
    for columns in weather[['tmax', 'tmin', 'depart', 'dewpoint',
       'wetbulb', 'sunrise', 'sunset',
       'preciptotal', 'stnpressure', 'sealevel', 'resultspeed',
       'resultdir', 'avgspeed']]:
        for i in range(1, len(weather)):
            if np.isnan(weather.loc[i, columns]) == False:
                weather.loc[i, columns]
            elif np.isnan(weather.loc[i, columns]) == True:
                weather.loc[i, columns] = weather.loc[i-1, columns]
```

```python
fill_prev()
```

```python
weather.describe()
```

### `tavg` 처리

```python
for i in range(len(weather)):
    if np.isnan(weather.loc[i, 'tavg']) == True :
        weather.loc[i, 'tavg'] = ((weather.loc[i, 'tmax'] + weather.loc[i, 'tmin']) / 2).round()
```

```python
print(len(weather[np.isnan(weather['tavg']) == True]))
weather[np.isnan(weather['tavg']) == True]
```

### `heat`, `cool` 처리

```python
weather[weather['station_nbr'].isin([2])][['heat', 'cool']].describe()
```

```python
for i in range(len(weather)):
    if np.isnan(weather.loc[i, 'heat']) == True :
        weather.loc[i, 'heat'] = (65 - weather.loc[i, 'tavg']).round()
```

```python
print(len(weather[np.isnan(weather['heat']) == True]))
weather[np.isnan(weather['heat']) == True]
```

```python
for i in range(len(weather)):
    if np.isnan(weather.loc[i, 'cool']) == True :
        weather.loc[i, 'cool'] = (weather.loc[i, 'tavg'] - 65).round()
```

```python
print(len(weather[np.isnan(weather['cool']) == True]))
weather[np.isnan(weather['cool']) == True]
```

### `snowfall`처리

```python
def fill_snowfall():
    for i in range(len(weather)):
        if np.isnan(weather.loc[i, 'snowfall']) == True:
            weather.loc[i, 'snowfall'] = weather.loc[i, 'preciptotal']
        elif np.isnan(i) == False:
            pass
```

### Station_5의 `sunrise` `sunset` 전 날을 이용해서 채우기

```python
def fill_5_sun():
    for columns in station_5[['sunrise', 'sunset']]:
        for i in range(len(station_5)):
            if station_5.loc[i, columns] != '-':
                station_5.loc[i, columns]
            elif station_5.loc[i, columns] == '-':
                station_5.loc[i, columns] = station_5.loc[i-1, columns]
```

```python
fill_5_sun()
```

```python
station_5[['sunrise', 'sunset']].head()
```

### 실수형 데이터로 채워넣기

```python
station_5['sunrise'] = station_5['sunrise'].apply(converge_sun)
station_5['sunset'] = station_5['sunset'].apply(converge_sun)
```

### null value 채워넣기

```python
for i in station_5.columns:
    station_5[i] = station_5[i].replace('-', np.nan)
    station_5[i] = station_5[i].replace(' ', np.nan)
    station_5[i] = station_5[i].replace('M', np.nan)
    station_5[i] = station_5[i].replace('T', np.nan)
    station_5[i] = station_5[i].replace('  T', np.nan)
```

### float로 전환

```python
for col in station_5[['tmax','tmin', 'tavg', 'depart', 'dewpoint', 'wetbulb', 'heat', 'cool', 
                 'snowfall', 'preciptotal', 'stnpressure','sealevel',
                 'resultspeed', 'resultdir', 'avgspeed']]:
    station_5[col] = station_5[col].astype(np.float64)
    
station_5 = pd.DataFrame(station_5, columns=['station_nbr', 'date','tmax', 'tmin', 'tavg', 'depart', 'dewpoint', 'wetbulb', 'heat',
       'cool', 'sunrise', 'sunset', 'codesum','codsum_YN', 'snowfall', 'preciptotal',
       'stnpressure', 'sealevel', 'resultspeed', 'resultdir', 'avgspeed'])
```

```python
def fillin_5(min_temp, max_temp):
    
    weather_temp = weather[(weather['sunrise'] >= min_temp) & (weather['sunrise'] < max_temp)]
    
    for i in range(len(station_5)):
        for column in station_5[['tmax', 'tmin', 'tavg', 'depart', 'dewpoint', 'wetbulb', 'heat',
       'cool', 'snowfall', 'preciptotal', 'stnpressure', 'sealevel', 'resultspeed', 'resultdir', 'avgspeed']]:
            mean_value = weather_temp[column].mean()
            if station_5.loc[i, 'sunrise'] >= min_temp and station_5.loc[i, 'sunrise'] < max_temp:
                if np.isnan(station_5.loc[i, column]) == True:
                    station_5.loc[i, column] = mean_value
```

```python
fillin_5(240, 270)
fillin_5(270, 300)
fillin_5(300, 330)
fillin_5(330, 360)
fillin_5(360, 390)
fillin_5(390, 420)
fillin_5(420, 450)
```
