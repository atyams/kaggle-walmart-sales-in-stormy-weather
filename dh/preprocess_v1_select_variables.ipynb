```python
data[data['sunrise'].isin(['-'])] = np.nan
```

```python
data = data.dropna()
data.reset_index(drop = True)
```

```python
def tmp_convert(sunrise):
    return float(sunrise[:2]) * 60 + float(sunrise[2:4])
```

```python
data['sunrise'] = data['sunrise'].apply(tmp_convert)
```

```python
data['sunset'] = data['sunset'].apply(tmp_convert)
```

```python
del data['day']
del data['month']
del data['station_nbr']
del data['year']
del data['codesum']
del data['date']
```
