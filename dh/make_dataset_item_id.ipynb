```python
tmp_df = pd.merge(weather,key)
tmp_df_1 = pd.merge(tmp_df,train)
tmp_df_1.tail()
```

```python
sale_info = pd.DataFrame(tmp_df_1, columns=['date', 'station_nbr', 'store_nbr', 'item_nbr', 'units'])
```

```python
def make_year(date):
    return int(date.split('-')[0])

def make_month(date):
    return int(date.split('-')[1])

def make_day(date):
    return int(date.split('-')[2])
```

```python
sale_info['year'] = sale_info['date'].apply(make_year)
sale_info['month'] = sale_info['date'].apply(make_month)
sale_info['day'] = sale_info['date'].apply(make_day)
```

```python
store_info = pd.DataFrame(sale_info, columns=['date','store_nbr','item_nbr', 'units'])
```

```python
store_info.tail()
```

```python
store_info = store_info.groupby(['date','store_nbr','item_nbr']).agg('sum')
```

```python
store_info = store_info[store_info['units'] > 0]
```

```python
store_info.reset_index(inplace = True)
```

```python
store_info = pd.merge(store_info, weather)
```

```python
store_info = pd.DataFrame(store_info, columns=['date','station_nbr','store_nbr','item_nbr','depart','codesum'])
```
